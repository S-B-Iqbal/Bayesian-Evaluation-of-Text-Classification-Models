```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az
```

### Effect of Sample Size

```python
samples = 100_000
l = 0.5

# m0 = np.random.beta(a = 0+l, b=0+l, size=samples)
m1 = np.random.beta(a = 10+l, b=10+l, size=samples)
m2 = np.random.beta(a = 50+l, b=50+l, size=samples)
m3 = np.random.beta(a = 500+l, b=500+l, size=samples)
```

```python
plt.figure(figsize=(6,6.5))
sns.kdeplot(m1, label = r"$TP=10,FP=10$", color = 'purple', linewidth=3)
sns.kdeplot(m2,label = r"$TP=50,FP=50$", color = 'red', linewidth=3)
sns.kdeplot(m3,label = r"$TP=500,FP=500$", color = 'blue', linewidth=3)
plt.xlabel("Precision")
plt.ylabel("Density")
plt.xlim((0,1))
plt.legend(loc = 'upper left')
plt.show()
```

```python
p1 = np.random.beta(a = 200+l, b=200+l, size=samples)
p2 = np.random.beta(a = 30+l, b=20+l, size=samples)
```

```python
# fig, axs = plt.subplots(ncols=1)
plt.figure(figsize=(6,6.5))
sns.kdeplot(p1, label = r"$TP=200,FP=200$", color = 'black', linewidth=3)
sns.kdeplot(p2, label = r"$TP=30,FP=20$", color = 'blue', linewidth=3)
plt.xlim((0,1))
plt.xlabel("Precision")
plt.legend()
plt.show()
```
